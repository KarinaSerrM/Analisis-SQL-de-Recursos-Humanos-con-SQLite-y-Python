1. Mediante el uso de Python y SQLite genera un archivo de base de datos denominado “RH.db” que contenga una tabla bajo el nombre de “Detalle” que importe la información proveniente del archivo CSV.

```python
# Importamos las librerias a usar
import sqlite3
import os
import csv
```

```python
# Generampos la conexión
conn = sqlite3.connect("RH.db")
# Definimos nuestra variable de cursor para realizar nuestras operaciones
cur = conn.cursor()
```

```python
# Lo direcciono a donde tengo el archivo
os.chdir("/Users/karin")
#definimos instrucciones para abrir el archivo
with open('recursos_humanos.csv') as f:
    reader = csv.reader(f)
    # Definimos la variable de datos que sea la variable Real
    data = list(reader)
```

```python
# Definimos nuestra base de da tos
cur.execute("""CREATE TABLE IF NOT EXISTs Detalle(
               satisfaction_level REAL,
               last_evaluation REAL,
               number_project INTEGER,
               average_monthly_hours INTEGER,
               time_spend_company INTEGER,
               work_accident INTEGER,
               left INTEGER,
               promotion_last_5years INTEGER,
               sales TEXT,
               salary TEXT
               )
""")
```

```python
# Insertamos los datos en la tabla a través de un ciclo
for renglon in range(1, len(data)):
    cur.execute("INSERT INTO Detalle(satisfaction_level, last_evaluation, number_project, average_monthly_hours, time_spend_company, work_accident, left, promotion_last_5years, sales, salary) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)", data[renglon])
conn.commit()
```

2. Despliega los campos “sales”, “salary” y “satisfaction_level” para todos los registros de la base de datos de forma que los registros se encuentren ordenados de manera descendente de acuerdo al campo “satisfaction_level”.  

```python
cur.execute('SELECT sales, salary, satisfaction_level FROM Detalle ORDER BY satisfaction_level DESC')
result = cur.fetchall()
result
```

3. Visualización de los registros para los campos “salary”, “number_project” y “satisfaction_level” ordenados en ascendente por “number_project” y en descendente por  el campo “satisfaction_level”

```python
cur.execute('SELECT salary, number_project, satisfaction_level FROM Detalle ORDER BY number_project ASC, satisfaction_level DESC')
result = cur.fetchall()
# Incorporo titulos y dejo un espacio para cada columna a manera de que la información se vea más ordenada
print("salary        number_project    satisfaction_level")
print("------------------------------------------  ------")
for row in result:
    print(f"{row[0]:<15} {row[1]:<15} {row[2]:<15}")
```

4. Promedio obtenido de la última evaluación (campo “last_evaluation”) para cada departamento (campo “sales”)

```python
# Adocinalmente lo ordenamos por promedio para mejorar la comprención de los datos
cur.execute('SELECT sales, ROUND(AVG(last_evaluation), 3) AS promedio_evaluacion FROM Detalle GROUP BY sales ORDER BY  promedio_evaluacion DESC')
result = cur.fetchall()
# Incorporo titulos 
print("Departamento        promedio_evaluacion")
print("-------------------------------------------")
for row in result:
    print(f"{row[0]:<20} {row[1]:<15}")
```

5. Ahora se agrupan los resultados por el departamento (campo “sales”) y salario (campo “salary”)

```python
# Ahora tome el criterio de ordenarlos por departamento y salario
cur.execute('SELECT sales, salary, ROUND(AVG(last_evaluation), 3) AS promedio_evaluacion FROM Detalle GROUP BY sales, salary ORDER BY sales ASC, salary ASC')
result = cur.fetchall()
# Incorporo titulos 
print("Departamento    Salario          Promedio_evaluacion")
print("----------------------------------------------------")
for row in result:
    print(f"{row[0]:<15} {row[1]:<15} {row[2]:<15}")
```

6.1 Departamentos con promedio de accidentes de trabajo mayor a 0.15.

```python
# Ahora ocupo la triple comilla por el largo del código y adicionalmente lo ordeno por el promedio de accidentes
cur.execute("""
SELECT 
    sales,
    ROUND(AVG(Work_accident), 3) as  promedio_accidentes
FROM 
    Detalle
GROUP BY 
    sales
HAVING 
    AVG(Work_accident) > 0.15
ORDER BY 
    promedio_accidentes DESC
""")
result = cur.fetchall()

# Incorporo titulos 
print("Departamento        Promedio_accidentes")
print("---------------------------------------")
for row in result:
    print(f"{row[0]:<20} {row[1]:<15}")
```

6.2 Departamentos con más de 200 accidentes totales

```python
cur.execute("""
SELECT 
    sales,
    SUM(Work_accident) AS total_accidentes    
FROM 
    Detalle
GROUP BY 
    sales
HAVING 
    SUM(Work_accident) > 200
ORDER BY 
    total_accidentes DESC
""")
result = cur.fetchall()

# Incorporo titulos 
print("Departamento        Total_accidentes")
print("---------------------------------------")
for row in result:
    print(f"{row[0]:<20} {row[1]:<15}")
```
